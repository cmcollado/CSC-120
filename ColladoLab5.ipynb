```python
#Question 4
for i in range(len(names)):
  count = 0
  for j in range(len(names)):
    if len(names[i]) >= len(names[j]) and i!=j:
      count += 1
  print("Names equal to or shorter than " + names[i] + " are " + str(count))
```

```python
#Question 5
i = 0 
while i < len(names):
  count = 0
  j = 0 
  while j < len(names):
    if len(names[i]) >= len(names[j]) and i!=j:
      count += 1
    j += 1  
  print("Names equal to or shorter than " + names[i] + " are " + str(count))
  i += 1
```

```python
#Question 6
import random
top_scores = []

def leaderboard(list, num):
  if len(list) < 5:
    list.append(num)
  else:
    if num > list[0]:
      list.append(num)
      del list[0]
      print("-----New leaderboard score!-----")  

for i in range(50):
  rand_num = random.randint(0,100)
  print("New number generated: " + str(rand_num))
  leaderboard(top_scores, rand_num)
  top_scores = sorted(top_scores) 
  print(top_scores)
```

```python
#Extra Credit 1
def Bubble(list):
  print("Unsorted: " + str(list))
  for i in range(len(list) - 1):
    for j in range(len(list) - 1 - i):
      if list[j] > list[j+1]:
        a = list.pop(j)
        list.insert(j+1, a)
  print("Sorted: " + str(list))      

test = [7, 2, 5, 6, 3, 1, 4]

Bubble(test)
```
